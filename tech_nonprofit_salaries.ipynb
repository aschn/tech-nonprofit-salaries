```python
import pandas as pd
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
# get data from google sheets csv file
url = 'https://docs.google.com/spreadsheets/d/1u3aH86lFRL4pLb0qV2JQFzlmYd_TduhPlwuIwOv_UXk/pub?gid=0&single=true&output=csv'
df = pd.read_csv(url)
print 'data types:', df.columns
print 'orgs:', df['Org'].unique()
```

```python
def plot_by_year(df, job):
    # fit by org age, all data
    base_yr = 2010
    m_all, b_all = np.linalg.lstsq(np.array([df['Year']-base_yr, np.ones_like(df['Year'])]).T,
                                   df['Salary_1000s'])[0]
    xs = range(2007, 2015)
    ys_all = [(i-base_yr)*m_all + b_all for i in xs]

    # medians
    medians = df.groupby('Year').aggregate(np.median)['Salary_1000s']

    # plot
    plt.figure()
    for org, orggroup in df.groupby('Org'):
        plt.plot(orggroup['Year'], orggroup['Salary_1000s'], 'o-', color='0.75')
    plt.plot(xs, ys_all, label='fit: %.1f+%.1f*(yr after %d)' % (b_all, m_all, base_yr))
    plt.plot(medians.index, medians, 'r', label='medians')
    plt.xlabel('Year')
    plt.ylabel('Salary ($1000s)')
    plt.xlim(2006, 2015)
    plt.title('%s salaries by year' % job)
    plt.legend(loc='upper left')
    plt.ticklabel_format(useOffset=False)

for cat, group in grouped:
    plot_by_year(group, 'C%sO' % cat)
```

```python
def plot_by_org_age(df, job):
    # drop missing data
    cleaned = df.dropna()
    
    # fit by org age, all data
    m_all, b_all = np.linalg.lstsq(np.array([cleaned['Org_Age'], np.ones_like(cleaned['Org_Age'])]).T,
                                   cleaned['Salary_1000s'])[0]
    xs = range(1, 14)
    ys_all = [i*m_all + b_all for i in xs]

    # medians
    medians = cleaned.groupby('Org_Age').aggregate(np.median)['Salary_1000s']

    # plot
    plt.figure()
    for org, orggroup in cleaned.groupby('Org'):
        plt.plot(orggroup['Org_Age'], orggroup['Salary_1000s'], 'o-', color='0.75')
    plt.plot(xs, ys_all, label='fit: %.1f+%.1f*age' % (b_all, m_all))
    plt.plot(medians.index, medians, 'r', label='medians')
    plt.xlabel('Years in operation')
    plt.ylabel('Salary ($1000s)')
    plt.title('%s salaries by org age' % job)
    plt.legend(loc='lower right')
    
for cat, group in grouped:
    plot_by_org_age(group, 'C%sO' % cat)
```

```python
def compare_positions(df, catx, caty):
    # assemble
    grouped = df.groupby(['Org', 'Year'])
    pieces = []
    for name, group in grouped:
        group.index = group['Category']
        if catx in group['Category'] and caty in group['Category']:
            piece = [group.loc[catx]['Salary_1000s'], group.loc[caty]['Salary_1000s']]
            pieces.append(piece)
    vs = pd.DataFrame(pieces, columns=[catx, caty])
    
    # fit, all data
    m_all, b_all = np.linalg.lstsq(np.array([vs[catx], np.ones_like(vs[catx])]).T,
                                   vs[caty])[0]

    plt.figure()
    plt.plot(range(450), range(450), 'k', label='equal')
    plt.scatter(vs[catx], vs[caty], color='0.75')
    plt.plot(vs[catx], [x*m_all + b_all for x in vs[catx]], 'b',
             label='fit: %.1f + %.2f*C%sO' % (b_all, m_all, catx))
    plt.xlim([0, 450])
    plt.ylim([0, 450])
    plt.legend(loc='upper left')
    plt.xlabel('C%sO salary ($1000s)' % catx)
    plt.ylabel('C%sO salary ($1000s)' % caty)
    plt.title('C%sO vs C%sO salaries' % (caty, catx))

compare_positions(df, 'E', 'T')
compare_positions(df, 'E', 'F')
compare_positions(df, 'E', 'O')
```

```python
def raise_by_org_age(df, job):
    # drop missing data
    cleaned = df.dropna()
 
    # plot absolute
    plt.figure()
    plt.plot(range(15), np.zeros(15), 'k')
    for org, orggroup in cleaned.groupby('Org'):
        raises_abs = orggroup.sort('Org_Age')['Salary_1000s'].diff()
        plt.plot(orggroup['Org_Age'][:-1], raises_abs[1:], 'o-', color='0.75')
    plt.xlabel('Years in operation')
    plt.ylabel('Raise after year ($1000s)')
    plt.title('%s raises by org age (absolute)' % job)

    # plot percentage
    plt.figure()
    plt.plot(range(15), np.zeros(15), 'k')
    for org, orggroup in cleaned.groupby('Org'):
        raises_pct = orggroup.sort('Org_Age')['Salary_1000s'].diff() / orggroup.sort('Org_Age')['Salary_1000s'] * 100
        plt.plot(orggroup['Org_Age'][:-1], raises_pct[1:], 'o-', color='0.75')
    plt.xlabel('Years in operation')
    plt.ylabel('Raise after year (percentage)')
    plt.title('%s raises by org age (relative)' % job)

for cat, group in grouped:
    raise_by_org_age(group, 'C%sO' % cat)
```

```python
def plot_by_budget(df, job):
    plt.figure()
    for org, orggroup in df.groupby('Org'):
        plt.plot(orggroup['Revenue_Ms']-orggroup['Grants_paid_Ms'], orggroup['Salary_1000s'], 'o-', color='0.75')
    plt.xlim([0, 70])
    plt.xlabel('Org budget ($Ms)')
    plt.ylabel('Salary ($1000s)')
    plt.title('%s salary by org budget (revenue - grants given)' % job)

for cat, group in grouped:
    plot_by_budget(group, 'C%sO' % cat)
```
